# matplotlib scientific 

#### I used matplotlib for scientific Data to generate a subplots of 2 rows and 2 columns with an error rate ... 
#### I didn't focus on what the data means, it's just for the purpose of the visualization. 

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
df = pd.DataFrame({'Date':['28-02-2022', '28-03-2022', '23-05-2022'],'Entrée (EUT)': [25, 28, 33], 'Sortie C1': [20, 20, 42],"Sortie C2":[38,49,82],"Sortie C3":[42,68,97],"Sortie C4":[29,40,84]})
df.index = pd.to_datetime(df['Date'],format='%d-%m-%Y')
del df["Date"]
df
```

```python
error = 20

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(16,15))

#graph [0,0]
ax[0,0].scatter(x=df.index, y= df["Entrée (EUT)"], color='red', s=40, label='Entrée (EUT)')
ax[0,0].scatter(x=df.index, y =df["Sortie C1"], color='c', s=40, label='Sortie C1')
ax[0,0].set_xlabel("Date d'analyse", fontsize=14)
ax[0,0].set_ylabel('DCO (mg/ l O2)', fontsize=14)
ax[0,0].set_ylim(0, 140)
ax[0,0].axhline(y=20,color='gray',ls='--',label = 'LQ')
ax[0,0].axhline(y=125,color='red',ls='--',label = 'Limite de rejet')
ax[0,0].set_title("Colonne 1")
ax[0,0].legend()

#erreur graph [0,0]
y1 = np.array([25,28,33])
y2 = np.array([20,20,42])
ax[0,0].errorbar(data =df ,x = df.index, y ="Entrée (EUT)",yerr= y1*error/100.,fmt = 'none', capsize = 5, ecolor = 'red', zorder = 1)
ax[0,0].errorbar(data =df ,x = df.index, y ="Sortie C1", yerr =y2*error/100., fmt = 'none', capsize = 5, ecolor = 'c', zorder = 1)

#graph [0,1]
ax[0,1].scatter(x=df.index, y= df["Entrée (EUT)"], color='red', s=40, label='Entrée (EUT)')
ax[0,1].scatter(x=df.index, y =df["Sortie C2"], color='c', s=40, label='Sortie C2')
ax[0,1].set_xlabel("Date d'analyse", fontsize=14)
ax[0,1].set_ylabel('DCO (mg/ l O2)', fontsize=14)
ax[0,1].set_ylim(0, 140)
ax[0,1].axhline(y=20,color='gray',ls='--',label = 'LQ')
ax[0,1].axhline(y=125,color='red',ls='--',label = 'Limite de rejet')
ax[0,1].set_title("Colonne 2")
ax[0,1].legend()



y1 = np.array([25,28,33])
y2 = np.array([38,49,82])
ax[0,1].errorbar(data =df ,x = df.index, y ="Entrée (EUT)",yerr= y1*error/100.,fmt = 'none', capsize = 5, ecolor = 'red', zorder = 1)
ax[0,1].errorbar(data =df ,x = df.index, y ="Sortie C2", yerr =y2*error/100., fmt = 'none', capsize = 5, ecolor = 'c', zorder = 1)

#graph [1,0]
ax[1,0].scatter(x=df.index, y= df["Entrée (EUT)"], color='red', s=40, label='Entrée (EUT)')
ax[1,0].scatter(x=df.index, y =df["Sortie C3"], color='c', s=40, label='Sortie C3')
ax[1,0].set_xlabel("Date d'analyse", fontsize=14)
ax[1,0].set_ylabel('DCO (mg/ l O2)', fontsize=14)
ax[1,0].set_ylim(0, 140)
ax[1,0].axhline(y=20,color='gray',ls='--',label = 'LQ')
ax[1,0].axhline(y=125,color='red',ls='--',label = 'Limite de rejet')
ax[1,0].set_title("Colonne 3")
ax[1,0].legend(loc='center left')



y1 = np.array([25,28,33])
y2 = np.array([42,68,97])
ax[1,0].errorbar(data =df ,x = df.index, y ="Entrée (EUT)",yerr= y1*error/100.,fmt = 'none', capsize = 5, ecolor = 'red', zorder = 1)
ax[1,0].errorbar(data =df ,x = df.index, y ="Sortie C3", yerr =y2*error/100., fmt = 'none', capsize = 5, ecolor = 'c', zorder = 1)

#graph [1,1]
ax[1,1].scatter(x=df.index, y= df["Entrée (EUT)"], color='red', s=40, label='Entrée (EUT)')
ax[1,1].scatter(x=df.index, y =df["Sortie C4"], color='c', s=40, label='Sortie C4')
ax[1,1].set_xlabel("Date d'analyse", fontsize=14)
ax[1,1].set_ylabel('DCO (mg/ l O2)', fontsize=14)
ax[1,1].set_ylim(0, 140)
ax[1,1].axhline(y=20,color='gray',ls='--',label = 'LQ')
ax[1,1].axhline(y=125,color='red',ls='--',label = 'Limite de rejet')
ax[1,1].set_title("Colonne 4")
ax[1,1].legend()

#erreur graph [0,0]
y1 = np.array([25,28,33])
y2 = np.array([29,40,84])
ax[1,1].errorbar(data =df ,x = df.index, y ="Entrée (EUT)",yerr= y1*error/100.,fmt = 'none', capsize = 5, ecolor = 'red', zorder = 1)
ax[1,1].errorbar(data =df ,x = df.index, y ="Sortie C4", yerr =y2*error/100., fmt = 'none', capsize = 5, ecolor = 'c', zorder = 1)

fig.savefig('C:/Users/Kemel/Desktop/ikram/DCO.png',dpi=1200)
```
