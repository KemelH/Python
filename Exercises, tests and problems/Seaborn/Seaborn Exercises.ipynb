## The Data
Visualization of the data with seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic') #import the dataframe of titanic 
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter') #joinplot of the columns x =fare, y=age
```

```python
sns.displot(titanic['fare'],bins=30,kde=False,color='red') #Displot of the column 'fare'
```

```python
sns.boxplot(x ='class',y ='age',data=titanic) #boxplot of the columns x='class', y="age"
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='magma') #swarmplot of the columns x='class' , y ='age'
```

```python
sns.countplot(x='sex',data=titanic) #countplot of the column 'sex'
```

```python
sns.heatmap(titanic.corr(),cmap='magma') #heatmap of the dataset 
plt.title("Titanic corr")
```

```python
g = sns.FacetGrid(data=titanic,col='sex') #facetgrid of the column age by sex
g.map(plt.hist,'age')
```
