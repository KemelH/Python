# Pandas Data Visualization Exercise

```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline #to visualize chart and graphe on jupyter notebook
```

```python
df3.info()
```

```python
df3.head()
```

```python
df3.plot.scatter(x="a",y="b",color='red',figsize=(13,3),s=40) #scatter plot of b vs a
```

```python
plt.style.use("ggplot") #to use matplotlib with pandas
df3['a'].plot.hist(bins=25,alpha=0.6) #histogram of the column 'a'
```

```python
df3.boxplot(column=["a","b"]) #boxplot comparing the 'a' and 'b' columns
```

```python
df3["d"].plot.kde(ls='--') #Kde plot of the "d" column
```

```python
df3.loc[0:30].plot.area(alpha=0.4) #aera chart of all of data frame 
```

```python
x = df3.loc[0:30].plot.area(alpha=0.4)
x.legend(bbox_to_anchor=(1.0, 0.6)) #to display the legend outside of the plot 
```
