```python
import random 
import numpy as np
import numpy.random as rand
import copy
```

```python
class Tree:
    states=['.','T','F','B']

    def __init__(self,state=0):
        if state <0 or state >3:
            # throw excption
            pass
        self._state = state
        
    def display_tree(self):
        """
        return the state of the tree
        """
        return self.states[self.state]

    def burning_neighbor(self,burning_neighbor):
        """
        looking for burning neighbor and update the state of the stree accodring to the rules of propagation
        """
        if burning_neighbor and self._state==1:
            self._state=2
    
    @property
    def state(self):
        return self._state
    @state.setter 
    def state(self, x): 
        self._state = x 
```

```python
class Forest:
       
    def __init__(self,nCols=10,nRows=10,p=0.6):
        self.nCols=nCols
        self.nRows=nRows
        self.p = p #probability
        self.grid=[]
        self.init_grid()

    def init_grid(self):
        
        """
        a method to initialise the grid with a probability p for a tree in a cell
        """
        has_burning_tree=False
        for i in range(0,self.nCols):
            self.grid.append([])
            for j in range(0,self.nRows):
                state=0
                if rand.random() <= self.p: #if the random number is less than p so we impliment a tree in the case 
                    if not has_burning_tree and rand.random()<0.5 :
                        state=2
                        has_burning_tree=True
                    else:
                        state=1

                
                self.grid[i].append(Tree(state))



    def display(self):
        
        """
        a method to display the grid 
        """
        
        for i in range(0,self.nCols):
            line=''
            for j in range(0,self.nRows):
                line+=self.grid[i][j].display_tree()+' '


            print(line)
            
        
    def has_burning_neighbor(self,i,j):
        
        """
        a method that return true if there is a burning tree in the neighbourhood for a specific cell in the grid
        
        """
        return \
            (i-1>=0 and self.grid[i-1][j].state == 2) or \
            (i+1<self.nRows and self.grid[i+1][j].state) == 2 or \
            (j-1>=0 and self.grid[i][j-1].state == 2) or \
            (j+1<self.nCols and self.grid[i][j+1].state) == 2

        
    def update_grid(self): 
        
        """
        
        A method that updates the grid according to the rules of propagation
        
        """
        
        new_grid = copy.deepcopy(self.grid) #Coping the grid
        still_burning = False
        
        for i in range(0,self.nRows):
            for j in range(0,self.nCols):
                
                #if the cell was burning it became burnt
                if self.grid[i][j].state == 2: 
                    new_grid[i][j].state=3
                #if the cell was healthy and have a burning neighbour last turn it will burn 
                elif self.grid[i][j].state == 1 and self.has_burning_neighbor(i,j):
                    new_grid[i][j].state=2
                    still_burning = True
                
                    
        self.grid=new_grid
        return still_burning
        
    
    def proportion_of_trees(self):
        
        """
        Methode to return the proportion of tree in the grid
        """
        num_trees = 0
        for i in range(0,self.nRows):
            for j in range(0,self.nCols):
                if self.grid[i][j].state == 1:
                    num_trees += 1

        # Calculate the proportion of trees in the grid
        total_cells = self.nCols * self.nRows
        proportion = num_trees / total_cells

        return proportion
    
    def generate(self,n):
        
        """
        A method that takes as parameter a strictly positive integer 
        and that generates this number of generations according to the rules of propagation
        
        """
        still_burning = True
        self.display()
        
        for i in range(0,n):
            print("Generation n°"+str(i+1))
            still_burning = self.update_grid()
            self.display()
            print(self.proportion_of_trees())
            if not still_burning:
                print("No more propagation")
                break
                

    
```

```python
forest=Forest()
forest.generate(100)
```

```python
class toricForest(Forest):
        
    
    def has_burning_neighbor(self,i,j):
        
        """
        a method that return true if there is a burning tree in the neighbourhood for a specific cell in the grid
        we will asume that the grid is toric 
        """
        return \
            (self.grid[(i-1)%self.nRows][j].state == 2) or \
            (self.grid[(i+1)%self.nRows][j].state) == 2 or \
            (self.grid[i][(j-1)%self.nCols].state == 2) or \
            (self.grid[i][(j+1)%self.nCols].state) == 2
    
    
    
```

```python
forest = toricForest()
forest.generate(100)
```
